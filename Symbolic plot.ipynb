```python
from sympy import symbols
from sympy.plotting import plot
from sympy.plotting import plot3d
from sympy.plotting import plot3d_parametric_line
from sympy import plot_implicit
from sympy.plotting import plot3d_parametric_surface
from sympy import *
from sympy.plotting import PlotGrid
plt.style.use('dark_background')
x,v,y,u= symbols('x v y u')
p=plot(x**2,sin(x),title='hola',xlabel='x',ylabel='y',legend='True',xscale='linear',xlim=(-10,10),ylim=(-1,1),line_color='red')
#xscale='log'
```

```python
plot((x**2, (x, -6, 6)), (sin(x), (x, -5, 5)))
```

```python
plot_parametric(cos(u), sin(u), (u, -5, 5))
```

```python
 p1 = plot_implicit(Eq(x**2 + y**2, 5))
```

```python
plot_implicit(y > x**2)
```

```python
plot_implicit(x - 1, x_var=x)
```

## 3D plot

```python
plot3d(x*y, (x, -5, 5), (y, -5, 5))
```

```python
plot3d(x*y, -x*y, (x, -5, 5), (y, -5, 5))
```

```python
plot3d((x**2 + y**2, (x, -5, 5), (y, -5, 5)),(x+y,(x,-10,10),(y,-10,10)),surface_color='red')
```

```python
plot3d_parametric_line(cos(u), sin(u), u, (u, -5, 5))
```

```python
plot3d_parametric_surface(cos(u + v), sin(u - v), u - v,
    (u, -5, 5), (v, -5, 5),surface_color='red')
```
